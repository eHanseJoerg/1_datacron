# Chapter 5: Understanding the Data / NB1: Combine Airspace Openings

This notebook serves as a first practical step with the data. The goal is to create a SPARQL query that gets us the airspace definitions, the exact times when an airspace was opened, its capacity and the planned flights (the demand) into one Pandas DataFrame. This DataFrame will be saved and used in the notebook `5.2 Airspace Capacity Heatmap` as foundation for the visualization.

```python
import os.path
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import pandas as pd
import json
import cesiumpy
import random
import ast
from geomet import wkt
from pandas.io.json import json_normalize, read_json
from SPARQLWrapper import SPARQLWrapper, JSON, XML, RDF
from datetime import datetime
from IPython.display import HTML
from IPython.display import clear_output

#Set some parameters for nicer visualizations
pd.set_option('display.expand_frame_repr', False) #do not wrap the printout of Pandas DataFrames
pd.set_option('display.precision', 2)
mp.rcParams['figure.figsize'] = (12, 6)
mp.pyplot.style.use = False

# initialize my connection module which allows to connect oto both datAcron graph databases
from datacron_connector import TripleStoreConnector
ts107 = TripleStoreConnector(0)
ts109 = TripleStoreConnector(1)
```

# 1. Get all Airblocks of Spain as 3D coordinates --> df_airblocks

```python
airblock_file = 'data/allspanishairblocks.csv'
if os.path.isfile(airblock_file):
    df_airblocks = pd.read_csv(airblock_file)
else:
    qry = """
    PREFIX : <http://www.datacron-project.eu/datAcron#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX dul: <http://www.ontologydesignpatterns.org/ont/dul/DUL.owl#>
    PREFIX myfn: <java:datAcronTester.unipi.gr.sparql_functions.>

    SELECT  ?bigairspace ?airspace ?block ?wkt ?lowerlevel ?upperlevel 
    WHERE { 

       ?airspace dul:hasPart       ?block .
       ?block       :hasLowerLevel ?lowerlevel ;
                    :hasUpperLevel ?upperlevel ;
                    :hasGeometry   ?geom.
       ?geom        :hasMBR_WKT  ?wkt .

       OPTIONAL {?bigairspace dul:hasPart ?airspace.}.

       FILTER regex(str(?block), 'Airblock_LE', "i")
    }
    """
    df_airblocks = ts109.query(qry)
    df_airblocks = ts109.clean(df_airblocks)
    %time df_airblocks = df_airblocks.sort_values('block')
    df_airblocks.to_csv(airblock_file)

df_airblocks.sample()
```

# 2. Get all opening times of airblocks (only times, no coordinates) --> df_openings

Now, we check another "single object": the sector configurations that were enabled in the spanish airspace. The sector configurations describe which subset of sectors and airblocks was active at a specific time. Each airblock belongs to one or more specific configurations, as we will see later. Additionally, the source table also contains capacity values. A capacity of 999 means that no capacity limits are defined.

```python
openings_file = 'data/openings.csv'
if not os.path.isfile(openings_file):
    qry = """
    PREFIX : <http://www.datacron-project.eu/datAcron#>
    PREFIX dul: <http://www.ontologydesignpatterns.org/ont/dul/DUL.owl#>

    SELECT ?config ?airspace ?capacity ?start ?end
    WHERE {
      ?config    a                        :FM_Configuration ;
                 :hasCapacity             ?capacity ;
                 :configurationOfAirspace ?airspace ;
              dul:hasConstituent          ?period .
      ?period    :TimeStart               ?start ;
                 :TimeEnd                 ?end.

       FILTER regex(str(?config), 'AirspaceConfiguration_LE', "i")
    }
    ORDER BY ?start
    """
    df_openings = ts109.query(qry)
    df_openings = ts109.clean(df_openings)
    #Lets add a column that describes the duration of the openings:
    df_openings['duration'] = pd.to_timedelta(pd.to_datetime(df_openings['end']) - pd.to_datetime(df_openings['start']))   
    df_openings.to_csv(openings_file)
else:
    df_openings = pd.read_csv(openings_file, index_col=0)
    df_openings['duration'] = pd.to_timedelta(df_openings['duration'])
    

df_openings.sample()
```

# 3. Joining both queries: get 4D definitions of open airblocks (dfob)

By joining the first query on airblocks and the second query on active configurations, we can get the time, date and coordinates of all active airblocks in the past. 
This will later enable us to search for weather data at an airblocks specific position in space and time. Some results of former experiments and discussions have to be considered for the next query:

There was some simplification going on in the loading of the airblock coordinates into the triple store as well-known-text. According to Giorgos Santipantakis, I have to take care about the way of querying the data in order to distinguish between exact and simplified (boxed) geometries. I will make sure to get both geometries, as the simplified geometry may help in finding a center for "nearest weather" later on.

```python
oblock_file = 'data/open_airblocks.csv'
if not os.path.isfile(oblock_file):
    qry = """
    PREFIX : <http://www.datacron-project.eu/datAcron#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX dul: <http://www.ontologydesignpatterns.org/ont/dul/DUL.owl#>
    PREFIX bif: <java:datAcronTester.unipi.gr.sparql_functions.>

    SELECT  ?config ?airspace ?sector ?block 
            ?simpleWKT (bif:getGeom(?geom) as ?actualWKT) ?lowerlevel ?upperlevel
            ?start ?end ?capacity 
    WHERE { 

       # get blocks:
       ?sector dul:hasPart       ?block .
       ?block       :hasLowerLevel ?lowerlevel ;
                    :hasUpperLevel ?upperlevel ;
                    :hasGeometry   ?geom.
       ?geom        :hasMBR_WKT    ?simpleWKT .

       # get active config times for all these blocks:
       ?config      a              :FM_Configuration ;
                    :hasCapacity   ?capacity ;
                    :configurationOfAirspace ?airspace ;
                 dul:hasConstituent          ?period .
       ?period      :TimeStart               ?start ;
                    :TimeEnd                 ?end.
       
       ?airspace dul:hasPart ?sector.

       #filter for only spanish airblocks, and only April and May 2016:
       FILTER (regex(str(?block), 'Airblock_LE', "i") && str(?start)>"2016-04-01T00:00:00" && str(?end)<"2016-05-31T23:59:59").  

    }
    """
    dfob = ts109.query(qry)
    dfob = ts109.clean(dfob)
    print('qry done. Now augmenting data...')
    
    # After loading the data, we append it with some calculated data:    
    dfob['start'] = pd.to_datetime(dfob['start'])
    dfob['end'] = pd.to_datetime(dfob['end'])
    dfob['duration'] = dfob['end'] - dfob['start']
    dfob = ts109.wkt_to_geojson(dfob, 'simpleWKT', 'simpleJSON')
    dfob = ts109.wkt_to_geojson(dfob, 'actualWKT', 'actualJSON')
    dfob = ts109.center_coordinate(dfob, 'simpleJSON', 'centerPoint')

    # there is this error in the dataset that the end times do not include the last 59 seconds 
    # of the last minute of validity of the airspace.We correct this manually:
    dfob['add_s'] = pd.Timedelta('59 seconds')
    dfob['end'] = dfob['end'] +  dfob['add_s']
    dfob = dfob.drop('add_s', 1)                                       

    #assert datatypes:
    dfob['capacity'] = dfob['capacity'].apply(pd.to_numeric, errors='ignore', downcast='unsigned')
    dfob['lowerlevel'] = dfob['lowerlevel'].apply(pd.to_numeric, errors='ignore', downcast='float')
    dfob['upperlevel'] = dfob['upperlevel'].apply(pd.to_numeric, errors='ignore', downcast='float')
    #pd.to_numeric(dfob['demand'], errors='coerce', downcast='unsigned')
    
    #sort by column opening dates:
    dfob.sort_values(by='start', ascending=True, inplace=True)
    
    #save everything to disk:
    dfob.to_csv(oblock_file)
    
else:
    #File found. Reload it and convert / assert data types:
    print('File found. reading from filestore...')
    dfob = pd.read_csv(oblock_file, index_col=0)
    print('JSON to dict conversion...')
    dfob['actualJSON'] = dfob['actualJSON'].map(ast.literal_eval) #convert string to dict
    dfob['simpleJSON'] = dfob['simpleJSON'].map(ast.literal_eval)
    
    print('datatype conversion...')
    dfob['capacity'] = dfob['capacity'].apply(pd.to_numeric, errors='ignore', downcast='unsigned')
    dfob['lowerlevel'] = dfob['lowerlevel'].apply(pd.to_numeric, errors='ignore', downcast='float')
    dfob['upperlevel'] = dfob['upperlevel'].apply(pd.to_numeric, errors='ignore', downcast='float')
    dfob['start'] = pd.to_datetime(dfob['start'])
    dfob['end'] = pd.to_datetime(dfob['end'])
    dfob['duration'] = pd.to_timedelta(dfob['duration'])
    

dfob.sample(2)
```

## 5. Generating Capacity Values and a Demand/Capacity Ratio

The following things have to be considered when trying to create the ratio.
 - The definition of a capacity is number of airplanes passing through the sector ('bigairspace') per hour. 
 - An indicated capacity of '999' in the database means that no limit has been defined for this spatio-temporal region.
 - Third, as the sectors are built of several _?blocks_, and we work along airblocks, the sector capacity and the demand is shared among these airblocks. Some abstraction will be required to distribute capacity and demand over these blocks. 
 - Fourth, as of todays understanding, defining capacities and creating regulations is not the same thing. An airspace may be of limited capacity without being regulated. 
 - Fifth, for sector openings of more or less than one hour, the capacity has to be scaled accordingly to the length of the opening period.

As the capacity and demand are defined per sector, but we will visualize each airblock, the objects have different reference volumes. Hence, one of them has to be scaled to match the other one. But it is not possible to scale down the capacity to an airblock without huge 3D-geometric calculations or further domain knowledge. I decide to scale the demand, using the following abstraction: 
 - each airblock is considered to have the full capacity of the sector
 - each airblock is considered to be confronted with the total traffic demand for the sector

With this operation, we construct an abstraction that assumes that every demand element passes through all airblocks of a sector. This abstraction is closest to the raw data, where capacity/demand imbalances are calculated on a sector level, while enabling the visualization, because the geometries are defined on airblock levels.

The intention is now to split the airspace-openings into hourly intervals. Essentially, we are "stretching" the dataframe to be longer, by splitting up airspace openings that were longer than an hour into unique records with a duration of max. one hour. 
The capacity $c_{exact}$ can be derived from the hourly capacity $c_{hourly}$ by multiplying the hourly capacity with the opening duration of the airspace:
$$ c_{exact} = c_{airblock} * \frac{(t_{close} - t_{open})}{60} $$
 

### 5.1 Scaling for time: slicing the Opening Schemes into intervals of max. 1 hour

The following procedure will create a (huge) DataFrame by slicing long opening times into multiple rows with a duration of max. one hour. The capacity and the demand for this time slices is adjusted pro rata. Disclaimer: to be more exact, we could re-calculate demand queries on the now-sliced dataset, but for the demonstrating purposes, pro rata demand is sufficient.

As we already calculated the demand, we do not need the WKT definitions anymore. In order not to get `memory error`, we are going to throw away these columns before startign the slicing operation:

```python
# 1. Clean some columns for a smaller memory footprint (saves around 100MB)
dfob = dfob.drop('simpleWKT', 1)
dfob = dfob.drop('simpleJSON', 1)
dfob['old_duration'] = dfob['duration']
dfob['old_capacity'] = dfob['capacity']
```

Now want to slice the dataframe. For each row with duration of > 1 hour, we slice off an hour and repeat until no rows with duration > 1 hour are existent in the dataset. As we know. the longest opening times have been 1440 minutes, so we expect this while loop to run for a maximum of 24 times. 

```python
from IPython.display import clear_output
mytimedelta = pd.Timedelta('1 hour')

#Create a boolean mask with rows to split (a vector of (True, True, False, ...) of length len(df))
split_rows = (dfob['duration'] > mytimedelta)    
i = 1

while split_rows.any():
    #get new rows to append and adjust start time to 1 hour later.
    new_rows = dfob[split_rows].copy()
    new_rows['start'] = new_rows['start'] + mytimedelta

    #update the end time of old rows
    dfob.loc[split_rows, 'end'] = dfob.loc[split_rows, 'start'] + \
        pd.DateOffset(hours=1, seconds=-1)
    dfob = dfob.append(new_rows)
    
    #update the duration of all rows
    dfob['duration'] = dfob['end'] - dfob['start']
    
    #create an updated boolean mask
    split_rows = (dfob['duration'] > mytimedelta)
    
    #give info to user
    clear_output(wait=True)
    i = i + 1
    print('Iteration' + str(i) + ' done.')

dfob.sample(2)
    
```

### 5.2 Updating Capacity Values to the new time slices

```python
dfob.sort_index().reset_index(drop=True)
dfob['duration'] = dfob['end'] - dfob['start']    
dfob['capacity'] = dfob['capacity'].astype('float32')
dfob['capacity'] = dfob['capacity'] * dfob['duration'] / pd.Timedelta('1 hour')
dfob.sample()
```

## 6.2 Generate Demand Values by Examining Raw Data

Because the geospatial crossing at RDF store level does not work, we will examine the raw datasets provided by Fraunhofer IAIS. We will cross our opened airblock dataset `dfob` with the data in the file `/1_datAcron/data/FTFM_sector-events.csv`, which contains all sector entry and exit events for all flights of AIRAC cycle 411 (April 2016) and the complete EUROCONTROL region. As this CSV file has a size of 2.7 GB, it is necessary to filter and create a smaller file containing only the relevant data: the events of the spanish airspace. Also, as our basic `dfob` dataset contains 102000 rows, but demand and capacity are calculated on sector, not on airblock level, we can vastly reduce the iteration by iterating on sector level.

[Technical Comment]

Aforementioned reductions and consequent usage of Pandas vectorized operations reduced the time required to execute the following code from several days to a couple of minutes.

### 6.2.1 generate the demand file for each sector opening

```python
sectorfile = 'data/5.1.7-sector-openings.csv'
if not os.path.isfile(sectorfile):
    #Step 0: create a sector-level DataFrame to fill later on 
    dfde = dfob[['sector', 'start', 'end']].copy()          
    dfde = dfde.drop_duplicates()                           
    dfde['demand'] = 0 
    dfde.to_csv(sectorfile)
else:
    dfde = pd.read_csv(sectorfile, index_col=0)
    dfde['start'] = pd.to_datetime(dfde['start'])
    dfde['end'] = pd.to_datetime(dfde['end'])
```

```python
demandfile = 'data/5.1.7-generated-demand.csv'
if not os.path.isfile(demandfile):
    
    #No demand file found. We need to create the demand file from the sector events file. 
    #Step 1: read or create a smaller entry events file containing only spanish airblocks
    file = 'data/5.1.7-FTFM_sector-events_LE.csv'
    if not os.path.isfile(file):                       #Create a smaller event subset from the big file:
        chunksize = 50000
        targetFrames = []
        i = 0
        for dataFrame in pd.read_csv('data/5.1.7-FTFM_sector-events.csv', chunksize=chunksize, sep=';', 
                                    usecols=['FLIGHT_UID', 'IOBT', 'SECTOR_ID', 'EVENT_TYPE', 'DT', 'DURATION_SEC']):
            mask = (dataFrame['SECTOR_ID'].str.slice(0,2) == 'LE' )
            dff = dataFrame[mask].copy()
            if len(dff) > 0:
                targetFrames.append(dff) 
            clear_output(wait=True)
            i = i +1    
            print('Iteration ' + str(i) + ' done.')
        dfentries = pd.concat(targetFrames)
        dfentries.to_csv('data/5.1.7-FTFM_sector-events_LE.csv')
    else:
        print('file found or generated. Now reading from datastore...')
        dfentries = pd.read_csv('data/5.1.7-FTFM_sector-events_LE.csv')
        dfentries['DT'] = pd.to_datetime(dfentries['DT'], dayfirst=True)
        
    
    #Step 2: Creating Boolean Masks on Sectors to further reduce the entry event dataset in each loop
    print ('Now generating boolean masks for sectors...')
    dfsectornames = dfde['sector'].str.slice(9,-4).copy()
    dfsectornames = dfsectornames.drop_duplicates()
    masks = {}
    for elem in dfsectornames:
        mask = (dfentries['SECTOR_ID'] == elem)
        masks[elem] = mask
        
    
    #Step 3: Loop through sector openings (dfde) and entry events (dfentries) to get all flights
    #        that pass through the open sector
    print('Step 3: now iterating through rows...')
    i, n = 0, len(dfde)
    for index, row in dfde.iterrows():
        sector, start, end = row['sector'][9:-4], row['start'], row['end']            
        clear_output(wait=True)
    
        sect_mask = masks[sector]
        dfentries_1sect = dfentries[sect_mask]
        dfentries_1sect_1period = dfentries_1sect[((dfentries_1sect['DT'] >= start) & (dfentries_1sect['DT'] <= end) )]
        dfde.at[index, 'demand'] = dfentries_1sect_1period.shape[0]   
        print('Iteration ' + str(i) + ' of ' + str(n) + ': ' + str(sector) + ': ' + str(dfentries_1sect_1period.shape[0]) + ' flights found.')

    #Step 4: save to disk
    dfde.to_csv('data/5.1.7-generated-demand.csv')

else:
    dfde = pd.read_csv('data/5.1.7-generated-demand.csv', index_col=0)
    dfde['start'] = pd.to_datetime(dfde['start'])
    dfde['end'] = pd.to_datetime(dfde['end'])
    
    
dfde.head()
```

### 6.2.3 write the generated demand values back into the original `dfob` DataFrame:

Now we have the DataFrame `dfde`, also saved in 'data/5.1.7-generated-demand.csv', that contains the calculated demand for all the sector openings. This demand has to be fed back to the main `dfob` DataFrame. 

```python
dfob = pd.merge(dfob, dfde, on=['sector', 'start', 'end'], how = 'left', left_index=False, right_index=False )
dfob.head(9)
```
